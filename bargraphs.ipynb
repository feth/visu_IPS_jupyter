```python
import pandas as pd
```

```python
df = pd.read_json("/home/feth/Téléchargements/fr-en-ips-ecoles-ap2022.json")
```

```python
df
```

```python
sens = df.loc[df['code_insee_de_la_commune'] == "89387"]
```

```python
sens.at[23841, "nom_de_l_etablissment"] = 'ECOLE ELEMENTAIRE RIGAULT'
```

```python
sens = sens.sort_values(by="ips")
```

```python
colorsdict = {"public": 'b', "privé sous contrat": 'g'}
```

```python
colors = [colorsdict[cat_etab] for cat_etab in sens['secteur']]
```

```python
plot = sens.plot.bar(x="nom_de_l_etablissment", y="ips", color=colors, legend=False)
for p in plot.patches:
    plot.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.015))
plot.set_xlabel("nom de l'établissement")
plot.set_ylabel("IPS - indice de position sociale")
```
